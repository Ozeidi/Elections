```python
path='Election\spiders\output\consolidated.txt'
df=pd.read_csv(path,sep='\t',header=0)
df.drop('w',axis=1,inplace=True)
df.rename(columns={'wilaya_id':'w_id'},inplace=True)
def A(x):
    if  isinstance(x,str) :
        reshaped_text = arabic_reshaper.reshape(unicode(x,'utf8'))
        artext = get_display(reshaped_text)
        return artext
    elif isinstance(x,unicode):
        reshaped_text = arabic_reshaper.reshape(x)
        artext = get_display(reshaped_text)
        return artext
    return x
#df.Surname_neutral.apply(A)
df.head()
```
